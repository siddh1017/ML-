```python
import seaborn as sns
```

```python
df  = sns.load_dataset ('tips')
df.head()

# tip is dependent feature 
# sex to size are the independent features
```

```python
df.dtypes
```

```python
df.corr()  
# it shows the correlation between the float type fields in the frame
```

# Heatmap
*A correlation heatMap uses colored cells, to shown the 2D correlation matrix between 2 discrete dimensions.* 

```python
df.corr()
```

```python
sns.heatmap (df.corr())
```

# Join-Plot

*hepls to plot the dataFrame with 1 feature and 1 label (Baivariant)*

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'hex')
```

# Pair-Plot
*to plot the more than 1 features*

```python
sns.pairplot (df)
```

```python
# to include catogerical data like sex use hue=
sns.pairplot (df, hue = 'sex')
```

# Dist-Plot
*helps to check the distribution of columns feature using the histogram*

```python
sns.distplot (df['total_bill'])
```

```python
sns.distplot (df['tip'], bins = 10)   # the data will be plotted to range 10 (bins)
```
