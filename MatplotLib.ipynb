# MatplotLib Tutorial

* plotting library and its numerical extensions of NumPy.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange (1, 11)
y = np.arange (11, 21)
```

```python
a = np.arange (40, 50)
b = np.arange (50, 60)
```

# Scatter

```python
# plotting using the matplotlib 

# plt using scatter

plt.scatter (x, y, c= 'r')    # plots the points
plt.xlabel ('X-axis')    # label for x-axis
plt.ylabel ('Y-axis')    # label for y-axis 
plt.title ("2D-graph")   # title for the graph
plt.savefig ('image.png')   # save the graph in same working folder
```

```python
y = x*x
plt.plot (x, y,'r*-', c = 'g')   # plots the line 
```

# Plot 

```python
y = x*x
plt.plot (x, y,'r*-', linewidth = 2, markersize = 10)  # plots the line 
plt.xlabel ('X-axis')    # label for x-axis
plt.ylabel ('Y-axis')    # label for y-axis 
plt.title ("2D-graph")   # title for the graph
```

# Sub-Plot

```python
x = np.arange (0, 5*np.pi, 0.1)
y_sin = np.sin (x)
y_cos = np.cos (x)

plt.subplot (2, 1, 1)  
plt.plot (x, y_sin, 'r--')
plt.title ("sin wave")

plt.subplot (2, 1, 2)
plt.plot (x, y_cos, 'g--')
plt.title ("cos wave")

```

# Bar-Plot

```python
x1 = [2, 3, 10]
y1 = [1, 10, 2]

x2 = [1, 16, 4]
y2 = [23, 4, 15]

plt.bar(x1, y1)
plt.bar(x2, y2)

plt.xlabel ('X-axis')    # label for x-axis
plt.ylabel ('Y-axis')    # label for y-axis 
plt.title ("Bar-plot")   # title for the graph
```

# Histograms 

```python
a = np.array([12,34,5,24,64,23,63,54,24,10,88,43,47,77,33])
plt.hist(a)
plt.title ('histogram')
```

# BOX Plot 

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot (data, vert = True, patch_artist=True )
```

# PIE Chart

```python
labels = 'Python', 'C++', 'C', 'JAVA'
sizes = [220, 120, 245, 222] 
colors = ['gold', 'yellowgreen', 'red', 'lightskyblue']
explode = (0.1, 0, 0.1, 0)  # gaps

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)

```
