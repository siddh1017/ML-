**LOC AND ILOC  --- >
    --> LOC when we have to access the elements by the name of rows and colmmns
    --> ILOC when we have to access the elements by indexes of rows and colomns**

```python
newDf.loc[1, 'C']   # row 1 and cloumn 'C'
```

```python
newDf.loc[1,2] = 123   # this creates the extra column with head = 2
newDf
```

```python
newDf = newDf.drop (2, axis = 1)   # drops the column name = 2 whose along axis = 1 
```

```python
newDf
```

```python
newDf.loc[1:44, ['B', 'D']]    # prints the rows from 1 to 44 
                               # pritns the columns 'B' and  'D'
```

```python
newDf.loc [:, 'A' : 'E']  # All rows and 'C' to 'E' columns
```

```python
newDf.loc [(newDf['B'] < 0.3)]    
# prints the all rows with value of 'B' < 0.3 
```

```python
newDf.loc [(newDf['B'] < 0.3) & (newDf['D'] > 0.3)]    
```

```python
 newDf
```

```python
newDf.iloc [0, 4]     # 0th row and 4th colomn element
```

```python
newDf.iloc [:, 1:4]    # other properties are same loc
```
